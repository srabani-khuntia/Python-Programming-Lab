#### 1. Write a Python program to: Create a text file named file1.txt, Write the following content into the file: "Hello, welcome to Python file handling!", Read the content of the file and display it.

```python
with open("file1.txt", "w") as f:
    f.write("Hello, welcome to Python file handling!")
with open("file1.txt", "r") as f:
    print("Initial Content:", f.read())
```

#### 3. Create a Python script to: Count the number of lines, words, and characters in a given text file.

```python
with open("file1.txt", "r") as f:
    lines = f.readlines()
    line_count = len(lines)
    word_count = sum(len(line.split()) for line in lines)
    char_count = sum(len(line) for line in lines)
print(f"Lines: {line_count}, Words: {word_count}, Characters: {char_count}")
```

#### 4. Write a Python program to: Read a file line by line and store each line into a list, Display the list of lines, Print all 3 letter words, Print all 5 letter words starting with ap. 

```python
with open("file1.txt", "r") as f:
    lines_list = [line.strip() for line in f.readlines()]
print("Lines List:", lines_list)
all_words = " ".join(lines_list).split()
print("3-letter words:", [w for w in all_words if len(w) == 3])
print("5-letter words starting with 'ap':", [w for w in all_words if len(w) == 5 and w.lower().startswith('ap')])
```

#### 6. Write a Python script to: Copy the content of one file (source.txt) to another file (destination.txt)

```python
import shutil
source = "file1.txt"
destination = "destination.txt"
try:
    shutil.copy(source, destination)
    print(f"Successfully copied '{source}' to '{destination}'.")
except FileNotFoundError:
    print(f"Error: The file '{source}' was not found.")
```

#### 7. Write a Python program to: Merge the contents of two files (file1.txt and file2.txt) into a third file (merged.txt).

```python
files_to_merge = ["file1.txt", "file2.txt"]
output_file = "merged.txt"
print(f"Starting merge into '{output_file}'...")
with open(output_file, "w") as outfile:
    for fname in files_to_merge:
        try:
            with open(fname, "r") as infile:
                content = infile.read()
                outfile.write(content + "\n")
                print(f"  + Added content from '{fname}' ({len(content)} characters)")
        except FileNotFoundError:
            print(f"Skipping '{fname}': File does not exist.")
print("Merging process complete.")
```

#### 8. Create a program that: Reads a file and removes all blank lines. Save the updated content into a new file.

```python
input_file = "file1.txt"
output_file = "no_blanks.txt"
with open(input_file, "r") as f:
    lines = f.readlines()
clean_lines = [line for line in lines if line.strip()]
with open(output_file, "w") as f:
    f.writelines(clean_lines)
print(f"Removed blank lines. Total lines reduced from {len(lines)} to {len(clean_lines)}.")
print(f"Cleaned content saved to '{output_file}'.")
```

#### 9. Write a Python script to: Encrypt the content of a file by shifting each character by 3 positions (Caesar Cipher). Save the encrypted content into a new file. Save these lines into a separate file.

```python
source_file = "file1.txt"
encrypted_file = "encrypted.txt"
shift = 3
def encrypt(text, s):
    return "".join(chr(ord(char) + s) for char in text)
try:
    with open(source_file, "r") as f:
        original_text = f.read()
    encrypted_text = encrypt(original_text, shift)
    with open(encrypted_file, "w") as f:
        f.write(encrypted_text)
    print(f"Encryption complete! Shifted characters by {shift} positions.")
    print(f"Encrypted file saved as '{encrypted_file}'.")
except FileNotFoundError:
    print("Error: Source file not found for encryption.")
```

#### 10. Write a program to find the longest word in a file

```python
file_name = "file1.txt"
with open(file_name, "r") as f:
    words = f.read().split()
if words:
    longest_word = max(words, key=len)
    print(f"The longest word in the file is: '{longest_word}'")
    print(f"Length: {len(longest_word)} characters.")
else:
    print("The file is empty.")
```

#### 11. Write a program to count the occurrence of each word in a file.

```python
from collections import Counter
file_name = "file1.txt"
with open(file_name, "r") as f:
    words = f.read().lower().split()
word_counts = Counter(words)
print("Word Frequency Count:")
for word, count in word_counts.items():
    print(f" - {word}: {count}")
```

#### 12. Create a file Wordlist. txt. Write a program to read a five letter word which starts with ‘ap’ and ends with ‘e’

```python
file_name = "Wordlist.txt"
# apple apply apnea agree apace apoke apode ample april aphid
try:
    with open(file_name, "r") as f:
        words = f.read().split()
    matches = [w for w in words if len(w) == 5 and w.lower().startswith('ap') and w.lower().endswith('e')]
    print(f"Searching for 5-letter words starting with 'ap' and ending with 'e'...")
    if matches:
        print(f"ound matches: {list(set(matches))}")
    else:
        print("No matching words found.")
except FileNotFoundError:
    print(f"Error: '{file_name}' not found. Please create it first.")
```
