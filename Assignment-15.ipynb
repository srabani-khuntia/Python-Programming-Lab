#### 17. Create a DataFrame with columns: Roll No, Name, Marks1, Marks2, and Marks3. Perform the following operations:
#### 1. Calculate Total Marks and Percentage
#### 2. Add a new column Grade based on percentage
#### 3. Display students scoring more than 75%
#### 4. Find the student with second highest percentage
#### 5. Sort the DataFrame based on percenta

```python
data = {
    'Roll No': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks1': [85, 60, 90, 70],
    'Marks2': [80, 65, 95, 75],
    'Marks3': [90, 70, 85, 80]
}
df_students = pd.DataFrame(data)

# 1.
df_students['Total Marks'] = df_students[['Marks1', 'Marks2', 'Marks3']].sum(axis=1)
df_students['Percentage'] = (df_students['Total Marks'] / 300) * 100

# 2. 
def assign_grade(pct):
    if pct >= 80: return 'A'
    elif pct >= 60: return 'B'
    elif pct >= 40: return 'C'
    else: return 'F'
df_students['Grade'] = df_students['Percentage'].apply(assign_grade)

# 3. 
print("--- Students > 75% ---\n", df_students[df_students['Percentage'] > 75])

# 4.
second_highest = df_students.drop_duplicates(subset=['Percentage']).nlargest(2, \
                                                        'Percentage').iloc[-1]
print("\n2nd Highest Percentage: ", second_highest['Name'], "-", \
                                              second_highest['Percentage'], "%")

# 5. 
df_sorted = df_students.sort_values(by='Percentage', ascending=False)
print("\n--- Sorted by Percentage ---\n", df_sorted)
```

#### 18. Create a DataFrame with fields: EmpID, Name, Department, and Salary. Write a program to:
#### 1. Display employees of a specific department (user input) 
#### 2. Increase salary by 10% for all employees
#### 3. Find the highest and lowest salary
#### 4. Group employees by department and compute average salary 
#### 5. Save the DataFrame to a CSV file

```python
data = {
    'EmpID': [1, 2, 3, 4, 5],
    'Name': ['Eve', 'Frank', 'Grace', 'Heidi', 'Ivan'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [60000, 45000, 65000, 70000, 48000]
}
df_emp = pd.DataFrame(data)

# 1. 
user_dept = "IT" 
print(f"--- Employees in {user_dept} ---\n", df_emp[df_emp['Department'] == user_dept])

# 2. 
df_emp['Salary'] = df_emp['Salary'] * 1.10

# 3. 
print(f"\nHighest Salary: {df_emp['Salary'].max()}")
print(f"Lowest Salary: {df_emp['Salary'].min()}")

# 4. 
print("\n--- Avg Salary by Dept ---\n", df_emp.groupby('Department')['Salary'].mean())

# 5. 
df_emp.to_csv('employees_updated.csv', index=False)
print("\nData saved to 'employees_updated.csv'")
```

#### 19. Create a DataFrame student with columns: roll No, name, marks, and city, perform:
#### 1. Select rows where marks > 80
#### 2. Filter students from a specific city 
#### 3. Select specific columns using loc[] and iloc[] 
#### 4. Update marks of a particular student 
#### 5. Delete a column

```python
data = {
    'Roll No': [1, 2, 3, 4],
    'Name': ['Jack', 'Kate', 'Leo', 'Mia'],
    'Marks': [75, 85, 92, 60],
    'City': ['NY', 'LA', 'NY', 'Chicago']}
student_df = pd.DataFrame(data)

# 1. 
print("--- Marks > 80 ---\n", student_df[student_df['Marks'] > 80])

# 2. 
print("\n--- Students from NY ---\n", student_df[student_df['City'] == 'NY'])

# 3. 
print("\n--- Using loc (Name & Marks) ---\n", student_df.loc[:, ['Name', 'Marks']])
print("\n--- Using iloc (Columns index 1 and 2) ---\n", student_df.iloc[:, [1, 2]])

# 4.
student_df.loc[student_df['Name'] == 'Jack', 'Marks'] = 80
print("\n--- Jack's Marks Updated ---\n", student_df)

# 5. Delete a column (City)
student_df = student_df.drop('City', axis=1)
print("\n--- 'City' Column Deleted ---\n", student_df)
```

#### 20. A company stores product sales data with columns: ProductID, ProductName, Category, Price, and Quantity. Write a program to: 
#### 1. Calculate total revenue 
#### 2. Find top-selling products 
#### 3. Filter products with low sales 
#### 4. Group by category and compute total revenue 
#### 5. Export results to Excel

```python
data = {
    'ProductID': [101, 102, 103, 104],
    'ProductName': ['Laptop', 'Mouse', 'Desk', 'Chair'],
    'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture'],
    'Price': [1000, 25, 200, 150],
    'Quantity': [5, 50, 10, 8]
}
sales = pd.DataFrame(data)

# 1. 
sales['TotalRevenue'] = sales['Price'] * sales['Quantity']
overall_revenue = sales['TotalRevenue'].sum()
print(f"Overall Total Revenue: ${overall_revenue}\n")

# 2. 
print("--- Top 2 Selling Products ---\n", sales.nlargest(2, 'Quantity'))

# 3. 
print("\n--- Low Sales Products ---\n", sales[sales['Quantity'] < 10])

# 4. 
print("\n--- Revenue by Category ---\n", sales.groupby('Category')['TotalRevenue'].sum())

# 5. 
sales.to_excel('sales_report.xlsx', index=False)
print("\nResults exported to 'sales_report.xlsx'")
```

#### 21. A CSV file students.csv contains: RollNo, Name, Marks1, Marks2, and Marks3. Write a Python program to:
#### 1. Read the CSV file into a DataFrame 
#### 2. Calculate Total and Percentage
#### 3. Assign grades based on percentage 
#### 4. Display students scoring above 75%
#### 5. Find the student with second highest percentage
#### 6. Save the result into result.csv

```python
data = {'RollNo': [1,2,3], 
        'Name': ['A', 'B', 'C'], 
        'Marks1': [80, 60, 90], 
        'Marks2': [75, 65, 85], 
        'Marks3': [85, 70, 90]}
#1. 
df_csv = pd.DataFrame(data) 
print(df_csv, "\n")

# 2. 
df_csv['Total'] = df_csv[['Marks1', 'Marks2', 'Marks3']].sum(axis=1)
df_csv['Percentage'] = (df_csv['Total'] / 300) * 100
print(df_csv[['Name', 'Total', 'Percentage']], "\n")

# 3. 
df_csv['Grade'] = pd.cut(df_csv['Percentage'], 
                         bins=[0, 40, 60, 75, 100], 
                         labels=['F', 'C', 'B', 'A'])
print(df_csv[['Name', 'Percentage', 'Grade']], "\n")

# 4. 
print("--- Students > 75% ---\n", df_csv[df_csv['Percentage'] > 75])

# 5. 
if len(df_csv) >= 2:
    sec_high = df_csv.drop_duplicates(subset=['Percentage']).nlargest(2, \
                                                'Percentage').iloc[-1]
    print(f'Second Highest: {sec_high['Name']}, {sec_high['Percentage']:.2f}%')
else:
    print("Not enough students to find a second highest.")
print("\n")

# 6. 
df_csv.to_csv('result.csv', index=False)
print("Success! The final DataFrame has been saved to 'result.csv'.")
print(df_csv)
```
