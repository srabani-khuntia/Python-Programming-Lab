#### 1. Create a tuple with different data and create an empty tuple using built-in function.

#### 2. Print the 3rd element from beginning and 4th element from end of a tuple.

```python
t=('Srabani', 7, 1.1, True, '24BCSB03', 6, 2.6)
print(f'3rd element from beginning: {t[2]}\n4th element from the end: {t[-4]}')
```

#### 3. Count the number of items present in a tuple and then add two new items into it (one item using concatenation operator and another item using built-in function).

```python
t=('Srabani', 7, 1.1, True, '24BCSB03', 6, 2.6)
print('Number of items present in tuple: ', len(t))
t+=(19,)
print(t)
```

#### 4. Find the number of occurrences for any repeated item present in a tuple.

```python
t=('Srabani', 7, 1.1, True, 1.1, 1.1, '24BCSB03', 6, 2.6)
n=t.count(1.1)
print(f'Number of occurences of {1.1} is {n}')
```

#### 5. Display the elements present in a tuple which are at the even position.

```python
t=eval(input('Enter tuple: '))
print('Elements at even position are:', end=' ')
for i in range(len(t)):
    if (i+1) % 2 == 0:
        print(t[i], end=' ')
```

#### 6. Remove the elements of a tuple which are present in the odd index position.

```python
t=eval(input('Enter tuple: '))
t_updated=()
for i in range(len(t)):
    if i % 2 == 0:
        t_updated+=(t[i],)
t=t_updated
print('Updated tuple: ', t)
```

#### 7. Remove the odd number elements from a tuple.

```python
t=eval(input('Enter tuple: '))
t_updated=()
for i in range(len(t)):
    if t[i]%2==0:
        t_updated+=(t[i],)
t=t_updated
print('Updated tuple: ', t)
```

#### 8. Reverse the elements of a tuple by using and without using a built-in function.

```python
t=eval(input('Enter tuple: '))
t_rev1=t[::-1]
t_rev2=tuple(reversed(t))
print('Reversed without using buit-in function: ', t_rev1)
print('Reversed with using buit-in function: ', t_rev2)
```

#### 14. Display the largest number of a tuple without using the built-in function max(). The user need to input the values in tuple from the keyboard.

```python
t=eval(input('Enter tuple: '))
lar = t[0]
for i in t:
    if i > lar:
        lar=i
print(f"The largest number is: {lar}")
```
