#### 1. Add first 'n' natural numbers.

```python
n = int(input('Enter n: '))
if n<=0:
    print('Not a natural number.')
else:
    sum = 0
    for i in range (0,n+1):
        sum += i
    print(f'Sum of first {n} natural numbers is {sum}')
```

#### 2. Show the Fibonacci series up to 10

```python
a,b = 0,1
print('0 1', end=' ')
for i in range(3,11):
    c = a+b
    print(c, end=' ')
    a,b = b,c
```

#### 3. Find the factorial of a number

```python
n = int(input('Enter a number: '))
if n<0:
    print('Factorial of a negative number does not exist.')
else:
    f = 1
    for i in range(1,n+1):
        f *= i
    print(f'Factorial of {n} is {f}')
```

#### 4. Find the sum of squares of first 'n' natural numbers

```python
n = int(input('Enter a number: '))
if n<=0:
    print('Not a natural number.')
else:
    sum = 0
    for i in range(1,n+1):
        sum += i*i
    print(f'Sum of squares of first {n} natural numbers is {sum}')
```

#### 5. Show the multiplication table from 1 to 10

```python
for i in range(1,11):
    print(f'Multiplication Table of {i}: ')
    for j in range(1,11):
        print(f'{i} x {j} = {i*j}')
    print()
```

#### 6. Find the prime numbers from 2 to 10

```python
print('Prime Numbers :', end=' ')
for i in range(2, 11):
    flag = 0
    for j in range(2,int(i**0.5 + 1)):
        if i%j==0:
            flag = 1
            break
    if flag==0:
        print(i, end=' ')
```
