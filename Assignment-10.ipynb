#### 1. Define a class Point, which can be used to create two-dimensional points. Create three objects of Point class and check, these points are non collinear.

```python
class Point:
    def __init__(self, x, y):
        self.x=x
        self.y=y
    def slope(self):
        return self.y/self.x
    
p1=Point(3,2)
p2=Point(1,1)
p3=Point(5,5)

#to check collinearity
if p1.slope()==p2.slope()==p3.slope():
    print('Three points are collinear')
else:
    print('Three points are non collinear')
```

#### 2. Define a class Box, having basic data members and member function and create Box objects and compute the volume of the Box

```python
class Box:
    def __init__(self,x,y,z):
        self.x=x
        self.y=y
        self.z=z
    def volume(self):
        return self.x*self.y*self.z
b=Box(1,2,3)
print('Volume of the box is ', b.volume())
```

#### 3. Define a class ComplexNumber, placed the necessary data members and member functions to perform basic Complex number operation.

```python
class ComplexNumber:
    def __init__(self, real=0.0, imag=0.0):
        self.real = real
        self.imag = imag
    def add(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)
    def subtract(self, other):
        return ComplexNumber(self.real - other.real, self.imag - other.imag)
    def multiply(self, other):
        new_real = (self.real * other.real) - (self.imag * other.imag)
        new_imag = (self.real * other.imag) + (self.imag * other.real)
        return ComplexNumber(new_real, new_imag)
    def display(self):
        sign = '+' if self.imag >= 0 else '-'
        print(f"{self.real} {sign} {abs(self.imag)}i")
        
c1 = ComplexNumber(3, 4)
c2 = ComplexNumber(1, 2)
print("C1: ", end=""); c1.display()
print("C2: ", end=""); c2.display()
c3 = c1.add(c2)
print("Addition: ", end="")
c3.display()
c4 = c1.multiply(c2)
print("Multiplication: ", end="")
c4.display()
```

#### 4. Define a class Time having data members hour, minute and second. The class should have a parameterized constructor to initialize its data members. It should also have methods displayTime() to display the time in HH:MM: SS format and addTime() to add two-time objects.

```python
class Time:
    def __init__(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second
    def displayTime(self):
        print(f"{self.hour:02d}:{self.minute:02d}:{self.second:02d}")
    def addTime(self, other_time):
        total_seconds = self.second + other_time.second
        carry_minute = total_seconds // 60
        new_second = total_seconds % 60
        total_minutes = self.minute + other_time.minute + carry_minute
        carry_hour = total_minutes // 60
        new_minute = total_minutes % 60
        new_hour = self.hour + other_time.hour + carry_hour
        return Time(new_hour, new_minute, new_second)

t1 = Time(2, 45, 50)
t2 = Time(1, 20, 15)
print("Time 1: ", end="")
t1.displayTime()
print("Time 2: ", end="")
t2.displayTime()
t3 = t1.addTime(t2)
print("Total Time: ", end="")
t3.displayTime()
```

#### 5. Write a python program that will create class BankAccount.
#### Data Members: accontNumber, accHolderName, accBalance, accType
#### Member Functions: withdraw(), deposit()

```python
class BankAccount:
    def __init__(self, accountNumber, accHolderName, accBalance, accType):
        self.accountNumber = accountNumber
        self.accHolderName = accHolderName
        self.accBalance = accBalance
        self.accType = accType
    def deposit(self, amount):
        if amount > 0:
            self.accBalance += amount
            print(f"Deposited: ${amount:.2f}")
            self.displayBalance()
        else:
            print("Deposit amount must be positive.")
    def withdraw(self, amount):
        if amount > self.accBalance:
            print(f"Failed to withdraw ${amount:.2f}: Insufficient funds!")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.accBalance -= amount
            print(f"Withdrawn: ${amount:.2f}")
            self.displayBalance()
    def displayBalance(self):
        print(f"Current Balance for {self.accHolderName}: ${self.accBalance:.2f}")
    def displayAccountInfo(self):
        print(f"\nAccount Number: {self.accountNumber}")
        print(f"Holder Name: {self.accHolderName}")
        print(f"Account Type: {self.accType}")
        self.displayBalance()
        print("-" * 30)

my_account = BankAccount(accountNumber="10012345", accHolderName="John Doe", accBalance=500.0, accType="Savings")
my_account.displayAccountInfo()
my_account.deposit(150.0)
my_account.withdraw(100.0)
my_account.withdraw(600.0)
```
