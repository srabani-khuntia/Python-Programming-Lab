#### 1. WAP to display the patterns as follows

##### (i) Half Pyramid

```python
for i in range (1,7):
    print('* '*i)
```

##### (ii) Inverted half Pyramid

```python
for i in range (6,0,-1):
    print('* '*i)
```

##### (iii) Hollow inverted half Pyramid

```python
for i in range(6,0,-1):
    if i==6:
        print('* '*6)
    elif i==1:
        print('*')
    else:
        print('*' + ' '*(i*2-3) + '*')
```

##### (iv) Full Pyramid

```python
for i in range(0,6):
    print(' '*(5-i) + '* '*(i+1))    
```

##### (v) Inverted full Pyramid

```python
for i in range(6,0,-1):
    print(' '*(6-i) + '* '*(i))    
```

##### (vi) Hollow full Pyramid

```python
for i in range(0, 6):
    if i == 0:
        print(' '*(5 - i) + '*')
    elif i == 5:
        print('* '*6)
    else:
        print(' '*(5 - i) + '*' + ' '*(2*i-1) + '*')
```
