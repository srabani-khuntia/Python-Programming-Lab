#### 18. Write a Python program using NumPy to:
#### 1. Create a 2D array for marks of 6 students in 4 subjects (user input)
#### 2. Compute total and average marks for each student
#### 3. Display students scoring above 70%
#### 4. Find the student with the second highest total marks
#### 5. Display subject-wise highest marks

```python
# 1. User Input
marks_list = []
print("Enter marks for 6 students in 4 subjects (space-separated):")
for i in range(6):
    row = list(map(int, input(f"Student {i} marks: ").split()))
    marks_list.append(row)
marks_array = np.array(marks_list)

# 2. Compute total and average
totals = marks_array.sum(axis=1)
averages = marks_array.mean(axis=1)
print("\nTotals per student:", totals)
print("Averages per student:", averages)

# 3. Students scoring > 70%
above_70_indices = np.where(averages > 70)[0]
print("Students scoring above 70% (0-indexed):", above_70_indices)

# 4. Second highest total
sorted_totals = np.sort(np.unique(totals))
if len(sorted_totals) > 1:
    second_highest = sorted_totals[-2]
    second_highest_student = np.where(totals == second_highest)[0][0]
    print(f"Student {second_highest_student} has the second highest total: {second_highest}")
else:
    print("Not enough distinct totals to find a second highest.")

# 5. Subject-wise highest
subject_highest = marks_array.max(axis=0)
print("Subject-wise highest marks:", subject_highest)
```

#### 19.Write a program to:
#### 1. Input two matrices
#### 2. Perform matrix addition, subtraction, and multiplication
#### 3. Find transpose of both matrices
#### 4. Compute determinant and inverse (if possible)
#### 5. Display results neatly

```python
#1. 
print("Enter 2x2 Matrix A (row by row, space-separated):")
A = np.array([list(map(float, input("Row 1: ").split())), 
              list(map(float, input("Row 2: ").split()))])
print("Enter 2x2 Matrix B (row by row, space-separated):")
B = np.array([list(map(float, input("Row 1: ").split())), 
              list(map(float, input("Row 2: ").split()))])

#2. 
print("\nAddition:\n", A + B)
print("Subtraction:\n", A - B)
print("Multiplication (Dot Product):\n", np.dot(A, B))

#3. 
print("\nTranspose of A:\n", A.T)
print("Transpose of B:\n", B.T)

#4. 
print("\nDeterminant of A:", np.linalg.det(A))
print("Determinant of B:", np.linalg.det(B))
try:
    print("\nInverse of A:\n", np.linalg.inv(A))
except np.linalg.LinAlgError:
    print("\nMatrix A is singular, no inverse exists.")
try:
    print("Inverse of B:\n", np.linalg.inv(B))
except np.linalg.LinAlgError:
    print("Matrix B is singular, no inverse exists.")
```

#### 20.Write a program to:

#### 2. Reshape it into a 2D array

#### 4. Sort elements in ascending and descending order
#### 5. Find maximum, minimum, and their indices

```python
#1. 
n = int(input("Enter number of elements (must be an even number for a 2-row reshape): "))
elements = list(map(int, input(f"Enter {n} elements (space-separated): ").split()))
arr_1d = np.array(elements)

#2. 
rows = 2
cols = n // 2
arr_2d = arr_1d.reshape(rows, cols)
print("\nReshaped to 2D:\n", arr_2d)

#3. 
flattened = arr_2d.flatten()
print("Flattened array:\n", flattened)

#4. 
asc_sorted = np.sort(flattened)
desc_sorted = asc_sorted[::-1]
print("Ascending Sort:", asc_sorted)
print("Descending Sort:", desc_sorted)

#5. 
max_val = flattened.max()
min_val = flattened.min()
print(f"Max: {max_val} at index {np.argmax(flattened)}")
print(f"Min: {min_val} at index {np.argmin(flattened)}")
```

#### 21.Write a program to:
#### 1. Create coefficient matrix A and constant matrix B(User Inputs)
#### 2. Solve the system using NumPy
#### 3. Verify solution by substituting back
#### 4. Handle singular matrix case

```python
#### 1. 
print("Solving Ax = B for a 2-variable system.")
print("Enter Coefficient Matrix A (2x2):")
A = np.array([list(map(float, input("Row 1: ").split())), 
              list(map(float, input("Row 2: ").split()))])
print("Enter Constant Matrix B (2x1):")
B = np.array(list(map(float, input("Values: ").split())))

#2. 
try:
    x = np.linalg.solve(A, B)
    print("\nSolution (x, y):", x)
    
#3. 
    verification = np.allclose(np.dot(A, x), B)
    print("Solution verified by back substitution:", verification)
#4. 
except np.linalg.LinAlgError:
    print("\nSingular Matrix detected. The system has no unique solution.")
```

#### 24.A company stores monthly sales data in a NumPy array. Write a program to:
#### 1. Create a 2-D array with months and sales(user inputs)
#### 2. Calculate total yearly sales
#### 3. Find month with highest and lowest sales
#### 4. Compute average monthly sales
#### 5. Identify months with below-average sales

```python
#1. 
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales_list = []
print("Enter sales data for 12 months:")
for month in months:
    sales = float(input(f"{month} Sales: "))
    sales_list.append(sales)
sales_data = np.array([months, sales_list])
sales_values = sales_data[1, :].astype(float)

#2. 
total_sales = sales_values.sum()
print("\nTotal Yearly Sales:", total_sales)

#3. 
max_index = sales_values.argmax()
min_index = sales_values.argmin()
print(f"Month with Highest Sales: {months[max_index]} ({sales_values[max_index]})")
print(f"Month with Lowest Sales: {months[min_index]} ({sales_values[min_index]})")

#4. 
avg_sales = sales_values.mean()
print("Average Monthly Sales:", round(avg_sales, 2))

#5. 
below_avg_indices = np.where(sales_values < avg_sales)[0]
below_avg_months = [months[i] for i in below_avg_indices]
print("Months with below-average sales:", below_avg_months)
```
