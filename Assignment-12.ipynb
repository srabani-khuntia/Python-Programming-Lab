#### 1. Define a class Employee having basic data members like empName, empID and empBasicSal, with constructors and necessary member functions. Define a class Manager which is inherited from Employee class and having a data member bonus. Create object of the class Manager and access manager details. 

```python
class Employee:
    def __init__(self, empName, empID, empBasicSal):
        self.empName = empName
        self.empID = empID
        self.empBasicSal = empBasicSal
    def display_employee(self):
        print(f"ID: {self.empID} \nName: {self.empName} \nSalary: {self.empBasicSal}")
        
class Manager(Employee):
    def __init__(self, empName, empID, empBasicSal, bonus):
        super().__init__(empName, empID, empBasicSal)
        self.bonus = bonus
    def display_manager(self):
        self.display_employee()
        print(f"Bonus: {self.bonus} \nTotal: {self.empBasicSal + self.bonus}")
    

m1 = Manager("Alice", "M101", 50000, 10000)
m1.display_manager()
```

#### 2. Problem Statement: Design a Library Management System using inheritance. The library contains different types of items such as:  Books,  Magazines. All items share some common properties, but each item also has specific properties.
#### Base Class: LibraryItem ( item_id,  title,  author,  display_info())
#### Derived Classes (1. Book(pages, genre), 2. Magazine(issue_number, publisher))
#### Each derived class should inherit from LibraryItem and add its own attributes and methods. Define the code which perform the following operation.  Create a base class LibraryItem.,  Define constructor to initialize item_id, title, author.  Create a method display_info()  Create derived classes Book, Magazine  Add specific attributes in each child class.  Override display_info() method in child classes.  Create objects and display details.

```python
class LibraryItem:
    def __init__(self, item_id, title, author):
        self.item_id = item_id
        self.title = title
        self.author = author
    def display_info(self):
        print(f"[{self.item_id}] {self.title} by {self.author}")

class Book(LibraryItem):
    def __init__(self, item_id, title, author, pages, genre):
        super().__init__(item_id, title, author)
        self.pages = pages
        self.genre = genre
    def display_info(self):
        super().display_info()
        print(f"Type: Book | Pages: {self.pages} | Genre: {self.genre}")

class Magazine(LibraryItem):
    def __init__(self, item_id, title, author, issue_number, publisher):
        super().__init__(item_id, title, author)
        self.issue_number = issue_number
        self.publisher = publisher
    def display_info(self):
        super().display_info()
        print(f"Type: Magazine | Issue: {self.issue_number} | Publisher: {self.publisher}")

b1 = Book("B001", "The Hobbit", "J.R.R. Tolkien", 310, "Fantasy")
b1.display_info()
m1 = Magazine("M99", "National Geographic", "Various", 452, "NG Society")
m1.display_info()
```

#### 3. Define an abstract class named as Shape, having data members dimOne and dimTwo. Extend this class to create two concrete classes named as Rectangle and Triangle. Override the computeArea() method in the sub classes. Invoke the computeArea() method. Create the object of derived class access computeArea() methods. 

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, dimOne, dimTwo):
        self.dimOne = dimOne
        self.dimTwo = dimTwo
    @abstractmethod
    def computeArea(self):
        pass

class Rectangle(Shape):
    def computeArea(self):
        return self.dimOne * self.dimTwo

class Triangle(Shape):
    def computeArea(self):
        return 0.5 * self.dimOne * self.dimTwo

rect = Rectangle(10, 20)
print(f"Rectangle Area: {rect.computeArea()}")
tri = Triangle(10, 20)
print(f"Triangle Area: {tri.computeArea()}")
```

#### 4. Create an abstract class named PaymentGateway. The abstract class should contain the following abstract methods: pay(amount), refund(amount), transaction_details()
#### Create the following child classes: CreditCardPayment, DebitCardPayment, UPIPayment
#### Each subclass should implement all abstract methods. Demonstrate the implementation using objects of each payment method. 

```python
from abc import ABC, abstractmethod

class PaymentGateway(ABC):
    @abstractmethod
    def pay(self, amount): 
        pass
    @abstractmethod
    def refund(self, amount): 
        pass
    @abstractmethod
    def transaction_details(self): 
        pass

class CreditCardPayment(PaymentGateway):
    def pay(self, amount): 
        print(f"Paid ${amount} via Credit Card.")
    def refund(self, amount): 
        print(f"Refunded ${amount} to Credit Card.")
    def transaction_details(self): 
        print("Details: Credit Card ending in 1234.")

class DebitCardPayment(PaymentGateway):
    def pay(self, amount):
        print(f"Paid ${amount} via Debit Card.")
    def refund(self, amount): 
        print(f"Refunded ${amount} to Debit Card.")
    def transaction_details(self):  
        print("Details: Debit Card Transaction ID: 987654321.")
    
class UPIPayment(PaymentGateway):
    def pay(self, amount): 
        print(f"Paid ${amount} via UPI.")
    def refund(self, amount): 
        print(f"Refunded ${amount} to VPA.")
    def transaction_details(self): 
        print("Details: UPI Transaction ID: 987654321.")

payments = [CreditCardPayment(), DebitCardPayment(), UPIPayment()]
for p in payments:
    p.pay(100)
    p.transaction_details()
```
