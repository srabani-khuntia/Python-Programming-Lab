#### 1. Multiply the numbers present in a list and display the final product.

#### 2. Print the twin prime numbers between 2 to 100.

```python
def isprime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: 
            return False
    return True
    
def istwinprime(i,j):
    if abs(i - j) != 2:
        return False
    return isprime(i) and isprime(j)

for i in range (2,99):
    if istwinprime(i, i+2):
        print(f'({i}, {i+2})') 
```

#### 4. Accepts a string and counts the number of upper and lower case letters.

```python
def count_cases(text):
    upper = sum(1 for char in text if char.isupper())
    lower = sum(1 for char in text if char.islower())
    print(f"Uppercase count: {upper}, Lowercase count: {lower}")
count_cases('Hello Srabani Khuntia... HAHAHAHA')
```

#### 11. Find the reverse of a number using recursion.

```python
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n // 10, rev * 10 + n % 10)

num = 12345
print(f"Original: {num}")
print(f"Reversed: {reverse_number(num)}")
```

#### 16. Swap the values between two numbers. Create a menu for the following operations:
####    a. Function should use positional arguments.
####    b. Use variable arguments in the function.

```python
# a. Positional arguments
def swap_positional(a, b):
    return b, a

# b. Variable arguments
def swap_variable(*args):
    if len(args) >= 2:
        return args[1], args[0]

# c. Keyword arguments
def swap_keyword(**kwargs):
    keys = list(kwargs.keys())
    if len(keys) >= 2:
        return {keys[0]: kwargs[keys[1]], keys[1]: kwargs[keys[0]]}

while True:
    print("\n--- SWAP MENU ---")
    print("1. Swap using Positional Arguments")
    print("2. Swap using Variable Arguments")
    print("3. Swap using Keyword Arguments")
    print("4. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        x, y = 10, 20
        x, y = swap_positional(x, y)
        print(f"Swapped Positional: x={x}, y={y}")
        
    elif choice == '2':
        res = swap_variable(50, 100)
        print(f"Swapped Variable: {res}")
        
    elif choice == '3':
        res = swap_keyword(first=99, second=88)
        print(f"Swapped Keyword: {res}")
        
    elif choice == '4':
        print("Exiting...")
        break
    else:
        print("Invalid choice, please try again.")
```
