#### 12. Search an element present in a list using linear search.

```python
l = []
for i in range(int(input("Enter the number of elements: "))):
    l.append(int(input(f"Enter value {i+1} : ")))
n=int(input('Enter element to search: '))
for i in l:
    if i==n:
        print(f'{i} found at index {l.index(i)}')
        break
else:
    print(f'{n} not found!!!')
```

#### 13. Search an element of a list using binary search.

```python
l = []
for i in range(int(input("Enter the number of elements: "))):
    l.append(int(input(f"Enter value {i+1} : ")))
n=int(input('Enter element to search: '))
l.sort()
low,high=0,len(l)-1
while(low<=high):
    mid=(low+high)//2
    if n==l[mid]:
        print(f'{n} found at index {l.index(n)}')
        break
    elif (n<l[mid]):
        high=mid-1
    else:
        low=mid+1
else:
    print(f'{n} not found!!!')   
```

#### 15. Split the even and odd elements of a list into two different lists.

```python
l = []
for i in range(int(input("Enter the number of elements: "))):
    l.append(int(input(f"Enter value {i+1} : ")))
l_even, l_odd = [],[]
for i in l:
    if i%2==0:
        l_even.append(i)
    else:
        l_odd.append(i)
print('Odd List: ', l_odd)
print('Even List: ', l_even)
```

#### 16. Add two matrices

```python
a = [[1,2,3],
    [4,5,6]]
b = [[1,3,5],
    [2,1,3]]
c = [[0,0,0],
    [0,0,0]]
for i in range(len(a)):
    for j in range(len(a[i])):
        c[i][j] = a[i][j] + b[i][j]
print(c)
```
