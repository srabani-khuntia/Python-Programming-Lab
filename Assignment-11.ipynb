#### 5. Enter the age of a person and create a custom exception InvalidAge if the person is too young for a job or too old for the job otherwise, he is selected for the job.

#### 7. User’s input is assumed to be a formula that consists of a number, an operator and, another number separated by white space (eg. 2 + 3). Split the input using split(). If the input is valid then perform the calculation and display the result otherwise, raise a custom exception FormulaError if:
####  the input does not consist of three elements (i.e, two operands and one operator).
####  the second input is not an operator.
####  the two operands used consists of float values.
####  while performing division operation, the second operand used is zero

```python
class FormulaError(Exception): 
    pass

try:
    formula = input('Enter formula: ').split()
    if len(formula) != 3: #binary operation
        raise FormulaError("Input must have 3 elements.")
    n1, op, n2 = formula #unpacking
    if op not in ['+', '-', '*', '/']:
        raise FormulaError("Invalid operator.")
    if '.' in n1 or '.' in n2:#no floats
        raise FormulaError("Operands must be integers, not floats.")
    n1, n2 = int(n1), int(n2) #string to integer
    if op == '/' and n2 == 0: #avoiding division by zero
        raise FormulaError("Division by zero.")
    
    results = {'+': n1+n2, '-': n1-n2, '*': n1*n2, '/': n1/n2}
    print(f'Result: {results[op]}')
except FormulaError as e:
    print(f'FormulaError: {e}')
except ValueError:
    print('FormulaError: Operands must be valid numbers.')
```

#### 8. Take your own input and raise an exception TypeError when wrong type of parameter is passed to a function.

#### 9. Handle multiple exceptions by creating different custom exceptions for the following:
####  NegativeError exception if user enter a negative number as input.
####  Enter data to a list and handle OverflowList exception if the maximum size of list is reached.
####  Use default exception to handle any type of exceptions.

```python
class NegativeError(Exception): 
    pass
class OverflowList(Exception): 
    pass
try:
    x = int(input("Enter a positive number: "))
    if x < 0: 
        raise NegativeError("No negatives allowed.")

    limit = 3
    my_list = [1, 2, 3]
    if len(my_list) >= limit:
        raise OverflowList("List reached maximum size.")
        
except NegativeError as e:
    print(e)
except OverflowList as e:
    print(e)
except Exception as e:
    print(f"Default Exception: {e}")
```
