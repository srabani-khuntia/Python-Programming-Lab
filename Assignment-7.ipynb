#### 11.  Count and display frequency of a word that appears in a given string. [Hint: take an empty dictionary, use split() to identify each words of string.]

```python
d={}
s=input('Enter String: ')
l=s.split()
st=set(l)
for i in st:
    d[i]=l.count(i)
print(d)
```

#### 12. Create a dictionary whose keys are month names and whose values are the number of days in the corresponding months.
#### a) Ask the user to enter a month name and use the dictionary to tell how many days are there in that month.
#### b) Print all of the keys in alphabetical order.
#### c) Print all of the month with 31 days.

```python
months={'January':31, 'February':28, 'March':31, 'April':30, 'May':31, 'June':30, 'July':31, 'August':31, 'September':30, 'October':31, 'November':30, 'December':31}

#(a)
m=input('Enter month: ')
print(f'{m.title()} has {months.get(m.title())} days.')

#(b)
print('Keys in alphabetical Order: ', sorted(months.keys()))

#(c)
m31=[]
for i in months:
    if months[i]==31:
        m31.append(i)
print('Months with 31 days are: ', m31 )
```

#### 13. Repeatedly ask the user to enter a team name and how many games the team has won and how many they lost. Store this information in a dictionary whose keys are the team names and the values are list of the form [wins, losses]. Perform the following using this dictionary:
#### a) Enter any team name and print the team's winning percentage.
#### b) Create a list whose entries are the number of wins of each team.

```python
d={}
ans='y'
while(ans=='y'):
    name=input('Enter team name:')
    wins=int(input('Enter number of wins:'))
    losses=int(input('Enter number of losses:'))
    l=[wins, losses]
    d[name]=l
    ans=input('Anymore?(y/n):')

#(a)
t=input('Enter team name: ')
l=d.get(t)
win_pt=l[0]/sum(l) * 100
print(f'Wins % of team {t} is {win_pt}%')

#(b)
wins_list=[i[0] for i in d.values()] #(id.values())[0]
print('List of wins: ', wins_list)
```
